```python

#pdf extractor
import fitz #from PyMuPDF
import re
path_one="contracts/ccnl_commercio_terziario_distribuzione_e_servizi.pdf"
path_two="contracts/ccnl_metalmeccanico_industria_conflavoro.pdf"
path_three="contracts/CCNL-Operai-Agricoli-e-Florovivaisti-2022-2025.pdf"
pdf_one=fitz.open(path_one)
pdf_two=fitz.open(path_two)
pdf_three=fitz.open(path_three)
def clean_text_for_parsing(text):
    text = re.sub(r"\.{4,}", " ", text)
    text = re.sub(r"[ \t]+", " ", text)
    text = re.sub(r"\n{3,}", "\n\n", text)
    text = text.strip()
    return text
def clean_text_for_embeddings(text):
    text = clean_text_for_parsing(text)
    text = re.sub(r"\n+", " ", text)
    text = re.sub(r"\s+", " ", text)
    return text.strip()
    
def text_extractor(pdf, contract_name):
    dictionary= {"contract_name": contract_name, "text": []}
    for page_number, page in enumerate(pdf, start=1):
            
            dictionary["text"].append({"page_number": page_number, "content_for_parsing": clean_text_for_parsing(page.get_text()), "content_for_embeddings": clean_text_for_embeddings(page.get_text())})
    return dictionary
text_one=text_extractor(pdf_one, "ccnl_commercio_terziario_distribuzione_e_servizi")
text_two=text_extractor(pdf_two, "ccnl_metalmeccanico_industria_conflavoro")
text_three=text_extractor(pdf_three, "CCNL-Operai-Agricoli-e-Florovivaisti-2022-2025")
```
